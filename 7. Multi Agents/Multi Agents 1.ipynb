* what tables - identify the right table for the problem statement
* then you identify the columns - understand schema
* top 10 from individual tables
* Generate query
* Execution of the query

```python
from langchain.agents import create_agent

def get_tables():
    """Return list of tables in the database."""
    conn = sqlite3.connect("sales_data.db")
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [row[0] for row in cursor.fetchall()]
    conn.close()
    return tables

def get_schema(table_name: str):
    """Return schema of a given table."""
    print('table name', table_name)
    conn = sqlite3.connect("sales_data.db")
    cursor = conn.cursor()
    cursor.execute(f"PRAGMA table_info({table_name});")
    results = cursor.fetchall()
    conn.close()
    return results

def execute_query(query: str):
    """Execute arbitrary SQL query and return results."""
    print(query)
    conn = sqlite3.connect("sales_data.db")
    cursor = conn.cursor()
    cursor.execute(query)
    results = cursor.fetchall()
    conn.close()
    return results


sql_agent = create_agent(llm,
                     tools=[get_tables, get_schema, execute_query],
                     system_prompt="""
                     You are a sqlite database agent, you have tools like get tables, get schema and execute query, 
                     understand column level stats first before generating the query, you can check for some basic stats and distinct values arond the required columns before you generate a query
                     """
                     )
```

```python
response = sql_agent.invoke({
  'messages': ['give sum of total sales for category electronic']
})
print(response)
```

```python

def get_chunks_with_metadata(query, filename):
  """This function helps in getting the chunks from the rag db based on the query"""
  retriever = vector_store.as_retriever(
      search_type="similarity_score_threshold",
      search_kwargs={"k": 5, "score_threshold": 0.4},
  )
  retrieved_docs = retriever.invoke(query, filter={"file_path": filename})
  return ' '.join([docs.page_content for docs in retrieved_docs])

def get_filenames_of_policies(query):
  """Get the filename for the given policy, the query can take any 1 value from below, use NA if it does not match any list here

  term insurance
  variable employee benefit
  group gratuity brochure
  future secure plan
  NA
  
  """
  policy_file_references = {'future secure plan': '0132667023-HDFC-Life-Group-Unit-Linked-Future-Secure-Plan.pdf',
   'group gratuity brochure': 'HDFC-GROUP-GRATUITY-BROCHURE.pdf',
   'term insurance': 'HDFC-Life-Group-Term-Insurance-Plan.pdf',
   'variable employee benefit': 'HDFC-LIFE-Group-Variable-Employee-Benefit-Plan.pdf'}
  if query in policy_file_references.keys():
    return policy_file_references[query]
  else:
    return ''

agentic_rag_metadata = create_agent(llm, 
                     tools=[get_filenames_of_policies, get_chunks_with_metadata],
                     system_prompt="""you are a agentic rag agent, 
                     your goal is answer queries related to insurance policy document, 
                     if user asks anythin beyond this, you say 'not supported'
                     Tool call rules:
                     first use get_filenames_of_policies to idenityf policy document file name which can be used in  get_chunks_with_metadata
                     """)
```

```python
def database_agent(query):
  """Database based agent who is connected to a database, transforms natural languages questions to the sql, executes and provides the repsonse"""
  response = sql_agent.invoke({
  # 'messages': ['give sum of total sales for alice manager']
      'messages': [query]
  })
  return response['messages'][-1].content

def policy_agent(query):
  """Policy based agent, who can answer any questiosn related to different policies"""
  response = agentic_rag_metadata.invoke({
    # 'messages': ['what is the age eligibility and who should be contacted for Grievance']
    'messages': [query]
  })
  return response['messages'][-1].content
```

```python
supervisor_agent = create_agent(
  llm,
  tools = [database_agent, policy_agent],
  system_prompt = """you have access to database agent and policy agent tools, 
  for any transation or data analytical queries use database agent tool
  for any specific policy related queries use policy tool
  """
)
```
