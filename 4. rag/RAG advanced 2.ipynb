```python
import os
pdf_files = [file for file in os.listdir() if file.lower().endswith('.pdf')]
pdf_files
```

```python
from langchain_pymupdf4llm import PyMuPDF4LLMLoader
def pdf_document_loader(file_path):
  loader = PyMuPDF4LLMLoader(file_path)
  docs = loader.load()
  print(len(docs))
  return docs
```

```python
parsed_docs = []
for pdf_file in pdf_files:
  docs = pdf_document_loader(pdf_file)
  parsed_docs.extend(docs)
```

```python
import getpass
import os

os.environ["AZURE_OPENAI_API_KEY"] = "openai key"
os.environ["AZURE_OPENAI_ENDPOINT"] = "https://ai-engineering-vidvatta1.openai.azure.com/"
os.environ["AZURE_DEPLOYMENT"]="gpt-5.4-mini"  # or your deployment
os.environ["API_VERSION"]="2025-04-01-preview"

from langchain_openai import AzureChatOpenAI

llm = AzureChatOpenAI(
    azure_deployment=os.environ["AZURE_DEPLOYMENT"],  # or your deployment
    api_version=os.environ["API_VERSION"])

llm.invoke('hello')
```

```python
from langchain_openai import AzureOpenAIEmbeddings

embeddings = AzureOpenAIEmbeddings(
    model="text-embedding-3-large")

embed = embeddings.embed_query('hello')
len(embed)
```

```python
## Recursive Character Splitting

from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter_recursive = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
docs_splits = splitter_recursive.split_documents(parsed_docs)
print(f"Chunks with RecursiveCharacterTextSplitter: {len(docs_splits)}")
```

```python
pip install -qU langchain langchain-pinecone
```

```python
os.environ["PINECONE_API_KEY"] = 'pinecone key'

from pinecone import Pinecone

pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])

index = pc.Index('policies-docs')
```

```python
from langchain_pinecone import PineconeVectorStore

vector_store = PineconeVectorStore(index=index, embedding=embeddings)
```

```python
from uuid import uuid4

uuids = [str(uuid4()) for _ in range(len(docs_splits))]
vector_store.add_documents(documents=docs_splits, ids=uuids)
```

```python
vector_retriever = vector_store.as_retriever()

# vector_retriever.invoke('what is the maturity for group term policy') # with no metadata

vector_retriever.invoke('what is the maturity for group term policy', filter={"file_path": "HDFC-Life-Group-Term-Insurance-Plan.pdf"})
```

```python
## 100 documents
## 1000 user
```

```python
def blog_chatbot(user_query):
  docs = vector_retriever.invoke(user_query, k=4) # retrieval : user query -> vector -> similarity -> top k
  context = '\n\n'.join([f"file_name: {doc.metadata['file_path']} \n page_content:  {doc.page_content}"
                         for doc in docs])

  prompt = f"""
  You are a chatbot, you have to answer the user query answer it only based on the context, if you dont know
  do not hallucinate and say you dont know.

  context: {context}
  user query: {user_query}
  """       # augmentation
  return llm.invoke(prompt), docs, context
```

```python

query = 'eligibility age for group unit payment policy'
response,  docs, context = blog_chatbot(query)
```

```python
from langchain_community.retrievers import BM25Retriever
bm25_retriever = BM25Retriever.from_documents(docs_splits, k=4)

bm25_chunks = bm25_retriever.invoke(query)
```

```python
def get_hybrid_rag_answers(user_query):
  bm25_chunks = bm25_retriever.invoke(user_query)
  semantic_chunks = vector_retriever.invoke(user_query)
  combined_chunks = bm25_chunks + semantic_chunks

  context = ' '.join(list(set([chunk.page_content for chunk in combined_chunks]))) # merge chunks into single string
  prompt = f"""
  You are a AI assitant, below is the user query and context, answer the user query based on the context only
  user query : {user_query}
  context: {context}
  """  
  response = llm.invoke(prompt)
  return response

print(get_hybrid_rag_answers(query).content)
```
