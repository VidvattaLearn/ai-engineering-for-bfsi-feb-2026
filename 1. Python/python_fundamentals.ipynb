## Section 2: Variables

Variables are **named containers** that store values. Python uses **dynamic typing** — you don't declare types.

```python
# Creating variables — no type declaration needed!
name = "Rahul"
age = 25
salary = 850000.50
is_employed = True

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Salary: ₹{salary:,.2f}")
print(f"Employed: {is_employed}")
```

```python
# Dynamic typing — same variable can change type
x = 42
print(f"x = {x}, type = {type(x)}")

x = "now I'm a string"
print(f"x = {x}, type = {type(x)}")

x = [1, 2, 3]
print(f"x = {x}, type = {type(x)}")
```

```python
# Multiple assignment
a, b, c = 10, 20, 30
print(a, b, c)

# Swap values (Pythonic way!)
a, b = b, a
print(f"After swap: a={a}, b={b}")
```

```python
# Variable naming conventions (PEP 8)
user_name = "Priya"         # snake_case — Python standard
MAX_RETRIES = 3             # UPPER_CASE — constants
order_total = 2499.99       # descriptive names

# These are INVALID:
# 2nd_place = "silver"     # Can't start with number
# my-var = 10              # No hyphens
# class = "MBA"            # 'class' is a reserved keyword
```

## Section 3: Basic Data Types

Python has 4 basic (scalar) data types:
- **int** — whole numbers
- **float** — decimal numbers
- **str** — text
- **bool** — True/False

### 3.1 Integers (int)

```python
# Integers — whole numbers with no size limit!
order_count = 1523
items_in_cart = 3
temperature = -5
population = 1_400_000_000   # underscores for readability (1.4 billion)

print(f"Order count: {order_count}")
print(f"Population: {population:,}")   # formatted with commas
print(f"Type: {type(order_count)}")
```

```python
# Python integers have NO size limit (unlike C/Java)
big_number = 2 ** 100
print(f"2^100 = {big_number}")
print(f"Digits: {len(str(big_number))}")
```

### 3.2 Floats (float)

```python
# Floats — decimal numbers
price = 999.99
gst_rate = 0.18
pi = 3.14159
scientific = 2.5e6          # 2,500,000.0

print(f"Price: ₹{price}")
print(f"Scientific: {scientific}")
print(f"Type: {type(price)}")
```

```python
# ⚠️ The infamous float precision issue
print(f"0.1 + 0.2 = {0.1 + 0.2}")           # 0.30000000000000004
print(f"0.1 + 0.2 == 0.3? {0.1 + 0.2 == 0.3}")  # False!

# Fix: use round()
print(f"round(0.1 + 0.2, 1) == 0.3? {round(0.1 + 0.2, 1) == 0.3}")  # True
```

### 3.3 Strings (str)

```python
# Strings — text data enclosed in quotes
name = "Rahul"
city = 'Mumbai'
bio = """This is a
multi-line string.
Very useful for long text."""

print(name)
print(city)
print(bio)
```

```python
# String operations
name = "Rahul Sharma"

print(f"Length: {len(name)}")
print(f"Upper: {name.upper()}")
print(f"Lower: {name.lower()}")
print(f"Title: {name.title()}")
print(f"First char: {name[0]}")
print(f"Last char: {name[-1]}")
print(f"First 5: {name[:5]}")
print(f"Contains 'Sharma': {'Sharma' in name}")
```

```python
# f-strings — the modern way to format strings (Python 3.6+)
item = "MacBook Pro"
price = 199900
quantity = 2

# Real-world: invoice line item
print(f"Item: {item}")
print(f"Price: ₹{price:,}")              # comma formatting
print(f"Total: ₹{price * quantity:,}")   # expression inside f-string
print(f"GST (18%): ₹{price * quantity * 0.18:,.2f}")
```

```python
# Useful string methods for real-world data cleaning
raw_email = "  Rahul.Sharma@Gmail.COM  "

clean_email = raw_email.strip().lower()
print(f"Raw: '{raw_email}'")
print(f"Clean: '{clean_email}'")

# Split and join
csv_line = "Rahul,25,Mumbai,Engineer"
fields = csv_line.split(",")
print(f"Fields: {fields}")

# Replace
phone = "+91-9876-543-210"
digits_only = phone.replace("-", "").replace("+", "")
print(f"Digits: {digits_only}")
```

### 3.4 Booleans (bool)

```python
# Booleans — True or False
is_active = True
is_deleted = False

print(f"Active: {is_active}, Type: {type(is_active)}")

# Booleans from comparisons
age = 20
print(f"Can vote (>=18): {age >= 18}")
print(f"Is teenager (13-19): {13 <= age <= 19}")
```

```python
# Truthiness — what evaluates to True/False?
# Falsy values: False, 0, 0.0, "", [], {}, None
# Everything else is Truthy

print(f"bool(0) = {bool(0)}")
print(f"bool(1) = {bool(1)}")
print(f"bool('') = {bool('')}")
print(f"bool('hello') = {bool('hello')}")
print(f"bool([]) = {bool([])}")
print(f"bool([1,2]) = {bool([1,2])}")
print(f"bool(None) = {bool(None)}")
```

### 3.5 Type Conversion

```python
# Converting between types
# String → Number (common when reading user input or API data)
age_str = "25"
age_int = int(age_str)
print(f"'{age_str}' → {age_int} (type: {type(age_int).__name__})")

price_str = "999.99"
price_float = float(price_str)
print(f"'{price_str}' → {price_float} (type: {type(price_float).__name__})")

# Number → String (common for display/formatting)
total = 1500
total_str = str(total)
print(f"{total} → '{total_str}' (type: {type(total_str).__name__})")

# Float → Int (truncates, does NOT round!)
print(f"int(3.9) = {int(3.9)}")     # 3 (not 4!)
print(f"int(3.1) = {int(3.1)}")     # 3
print(f"round(3.9) = {round(3.9)}") # 4 (if you want rounding)
```

```python
# Real-world: Processing API response data
# APIs often return everything as strings
api_response = {
    "price": "2499.99",
    "quantity": "3",
    "in_stock": "true"
}

price = float(api_response["price"])
qty = int(api_response["quantity"])
in_stock = api_response["in_stock"].lower() == "true"

total = price * qty
print(f"Item total: ₹{total:,.2f}")
print(f"In stock: {in_stock}")
```

## Section 4: Operators

Operators perform operations on values. Python has arithmetic, comparison, logical, and assignment operators.

### 4.1 Arithmetic Operators

```python
# Arithmetic operators
a, b = 17, 5

print(f"{a} + {b} = {a + b}")     # Addition: 22
print(f"{a} - {b} = {a - b}")     # Subtraction: 12
print(f"{a} * {b} = {a * b}")     # Multiplication: 85
print(f"{a} / {b} = {a / b}")     # Division: 3.4 (always float)
print(f"{a} // {b} = {a // b}")   # Floor division: 3
print(f"{a} % {b} = {a % b}")     # Modulo: 2
print(f"{a} ** {b} = {a ** b}")   # Power: 1419857
```

```python
# Real-world: E-commerce price calculation
base_price = 1299
gst_rate = 0.18
discount_pct = 0.10
quantity = 2

subtotal = base_price * quantity
discount = subtotal * discount_pct
after_discount = subtotal - discount
gst = after_discount * gst_rate
grand_total = after_discount + gst

print(f"Subtotal:      ₹{subtotal:,.2f}")
print(f"Discount (10%): -₹{discount:,.2f}")
print(f"After discount: ₹{after_discount:,.2f}")
print(f"GST (18%):     +₹{gst:,.2f}")
print(f"Grand Total:    ₹{grand_total:,.2f}")
```

```python
# Modulo (%) — super useful for checking even/odd, cycling, etc.
for i in range(1, 11):
    parity = "even" if i % 2 == 0 else "odd"
    print(f"{i:2d} is {parity}")
```

### 4.2 Comparison Operators

```python
# Comparison operators return True or False
x, y = 10, 20

print(f"{x} == {y}: {x == y}")   # Equal
print(f"{x} != {y}: {x != y}")   # Not equal
print(f"{x} > {y}: {x > y}")     # Greater than
print(f"{x} < {y}: {x < y}")     # Less than
print(f"{x} >= {y}: {x >= y}")   # Greater or equal
print(f"{x} <= {y}: {x <= y}")   # Less or equal
```

```python
# Real-world: Eligibility checks
age = 22
credit_score = 750
annual_income = 600000

print(f"Can vote (age >= 18): {age >= 18}")
print(f"Good credit (>= 700): {credit_score >= 700}")
print(f"Income eligible (>= 5L): {annual_income >= 500000}")

# Python supports chained comparisons!
print(f"Age 18-65: {18 <= age <= 65}")
print(f"Credit 700-850: {700 <= credit_score <= 850}")
```

### 4.3 Logical Operators

```python
# Logical operators: and, or, not
print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"True or False: {True or False}")
print(f"False or False: {False or False}")
print(f"not True: {not True}")
print(f"not False: {not False}")
```

```python
# Real-world: Loan eligibility
age = 28
income = 800000
credit_score = 720
existing_loans = 1

# All conditions must be True
eligible = (
    21 <= age <= 60
    and income >= 500000
    and credit_score >= 700
    and existing_loans <= 2
)

print(f"Loan eligible: {eligible}")

# Discount: student OR senior citizen OR first-time buyer
is_student = False
is_senior = False
is_first_time = True

gets_discount = is_student or is_senior or is_first_time
print(f"Gets discount: {gets_discount}")
```

### 4.4 Assignment Operators

```python
# Assignment operators — shorthand for common operations
balance = 10000
print(f"Starting balance: ₹{balance}")

balance += 5000   # balance = balance + 5000
print(f"After deposit:    ₹{balance}")

balance -= 3000   # balance = balance - 3000
print(f"After withdrawal: ₹{balance}")

balance *= 1.05   # balance = balance * 1.05 (5% interest)
print(f"After interest:   ₹{balance:.2f}")

# Also available: //=, %=, **=
```

## Section 5: Conditional Statements

Conditionals let your code **make decisions** — run different code based on conditions.

### 5.1 Basic if Statement

```python
# Basic if statement
temperature = 38.5

if temperature > 37.2:
    print("⚠️ Fever detected!")
    print("Please consult a doctor.")
```

```python
# if-else
balance = 15000
withdrawal = 20000

if balance >= withdrawal:
    balance -= withdrawal
    print(f"✅ Withdrawn ₹{withdrawal}. New balance: ₹{balance}")
else:
    print(f"❌ Insufficient funds. Balance: ₹{balance}, Requested: ₹{withdrawal}")
```

### 5.2 if-elif-else

```python
# Real-world: Delivery fee calculator (like Swiggy/Zomato)
order_total = 350

if order_total >= 500:
    delivery_fee = 0
    message = "🎉 Free delivery!"
elif order_total >= 200:
    delivery_fee = 30
    message = f"Delivery fee: ₹{delivery_fee}"
elif order_total >= 100:
    delivery_fee = 50
    message = f"Delivery fee: ₹{delivery_fee}"
else:
    delivery_fee = 70
    message = f"Delivery fee: ₹{delivery_fee}. Add ₹{100 - order_total} more to reduce!"

final_total = order_total + delivery_fee
print(f"Order: ₹{order_total} | {message}")
print(f"Final total: ₹{final_total}")
```

```python
# Real-world: Tax slab calculator (simplified Indian tax)
income = 1200000  # ₹12 lakhs

if income <= 300000:
    tax = 0
elif income <= 600000:
    tax = (income - 300000) * 0.05
elif income <= 900000:
    tax = 15000 + (income - 600000) * 0.10
elif income <= 1200000:
    tax = 45000 + (income - 900000) * 0.15
elif income <= 1500000:
    tax = 90000 + (income - 1200000) * 0.20
else:
    tax = 150000 + (income - 1500000) * 0.30

print(f"Income: ₹{income:,}")
print(f"Tax: ₹{tax:,.2f}")
print(f"Effective rate: {tax/income*100:.1f}%")
```

### 5.3 Nested Conditions & Ternary Operator

```python
# Nested conditions — use sparingly
has_account = True
is_verified = True
balance = 5000
amount = 3000

if has_account:
    if is_verified:
        if balance >= amount:
            print(f"✅ Transaction of ₹{amount} approved")
        else:
            print("❌ Insufficient balance")
    else:
        print("❌ Please verify your account first")
else:
    print("❌ Account not found")
```

```python
# Better: Flatten with early returns / combined conditions
if not has_account:
    print("❌ Account not found")
elif not is_verified:
    print("❌ Please verify your account first")
elif balance < amount:
    print("❌ Insufficient balance")
else:
    print(f"✅ Transaction of ₹{amount} approved")
```

```python
# Ternary operator — one-line conditionals
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age {age}: {status}")

score = 72
result = "PASS" if score >= 40 else "FAIL"
print(f"Score {score}: {result}")

# Real-world: Stock alert
stock_change = -2.5
alert = f"📈 Up {stock_change}%" if stock_change > 0 else f"📉 Down {abs(stock_change)}%"
print(alert)
```

## Section 6: Loops

Loops automate repetitive tasks. Python has `for` loops (iterate over items) and `while` loops (repeat until condition).

### 6.1 for Loops

```python
# Loop through a list
fruits = ["apple", "banana", "mango", "guava"]
for fruit in fruits:
    print(f"I like {fruit}!")
```

```python
# Loop through a string
for char in "Python":
    print(f"Letter: {char}")
```

```python
# range() — generate number sequences
# range(stop): 0 to stop-1
# range(start, stop): start to stop-1
# range(start, stop, step): with custom step

print("range(5):")
for i in range(5):
    print(i, end=" ")

print("\n\nrange(1, 6):")
for i in range(1, 6):
    print(i, end=" ")

print("\n\nrange(0, 20, 3):")
for i in range(0, 20, 3):
    print(i, end=" ")

print("\n\nCountdown:")
for i in range(5, 0, -1):
    print(i, end=" ")
print("🚀 Launch!")
```

```python
# enumerate() — get index AND value
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]

print("Product Catalog:")
for index, product in enumerate(products, start=1):
    print(f"  {index}. {product}")
```

```python
# Real-world: Process a list of transactions
transactions = [
    ("TXN001", 2500, "credit"),
    ("TXN002", 1200, "debit"),
    ("TXN003", 500, "credit"),
    ("TXN004", 3000, "debit"),
    ("TXN005", 750, "credit"),
]

total_credit = 0
total_debit = 0

for txn_id, amount, txn_type in transactions:
    if txn_type == "credit":
        total_credit += amount
        print(f"  ✅ {txn_id}: +₹{amount}")
    else:
        total_debit += amount
        print(f"  🔴 {txn_id}: -₹{amount}")

print(f"\nTotal Credits: ₹{total_credit}")
print(f"Total Debits:  ₹{total_debit}")
print(f"Net:           ₹{total_credit - total_debit}")
```

### 6.2 while Loops

```python
# while loop — repeat until condition is False
count = 5
while count > 0:
    print(f"Countdown: {count}")
    count -= 1
print("🚀 Liftoff!")
```

```python
# Real-world: Simulate API retry logic
import random

max_retries = 5
attempt = 0
success = False

while attempt < max_retries and not success:
    attempt += 1
    # Simulate 40% success rate
    success = random.random() < 0.4
    status = "✅ Success" if success else "❌ Failed"
    print(f"  Attempt {attempt}/{max_retries}: {status}")

if success:
    print(f"\nAPI call succeeded on attempt {attempt}!")
else:
    print(f"\nAll {max_retries} attempts failed. Giving up.")
```

### 6.3 break and continue

```python
# break — exit loop immediately
# Real-world: Find first out-of-stock item
inventory = [
    ("Laptop", 15),
    ("Mouse", 230),
    ("Webcam", 0),
    ("Keyboard", 45),
    ("Monitor", 0),
]

print("Checking inventory...")
for item, stock in inventory:
    if stock == 0:
        print(f"⚠️ ALERT: {item} is out of stock!")
        break
    print(f"  ✅ {item}: {stock} in stock")
```

```python
# continue — skip current iteration
# Real-world: Process only active orders
orders = [
    ("ORD001", "shipped", 2500),
    ("ORD002", "cancelled", 1800),
    ("ORD003", "shipped", 3200),
    ("ORD004", "cancelled", 950),
    ("ORD005", "delivered", 4100),
]

active_revenue = 0
print("Processing active orders:")
for order_id, status, amount in orders:
    if status == "cancelled":
        continue  # Skip cancelled orders
    active_revenue += amount
    print(f"  📦 {order_id} ({status}): ₹{amount}")

print(f"\nActive revenue: ₹{active_revenue:,}")
```

### 6.4 Nested Loops

```python
# Nested loops: Multiplication table
print("Multiplication Table (1-5):")
print("-" * 35)
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:4d}", end="")
    print()  # New line after each row
```

```python
# Real-world: Generate product-size combinations
products = ["T-Shirt", "Polo", "Hoodie"]
sizes = ["S", "M", "L", "XL"]

print("SKU Catalog:")
sku_count = 0
for product in products:
    for size in sizes:
        sku_count += 1
        sku = f"{product[:3].upper()}-{size}"
        print(f"  SKU-{sku_count:03d}: {sku} ({product} - {size})")

print(f"\nTotal SKUs: {sku_count}")
```

```python
# ANSWER 1: Bill Splitter
bill_amount = 2400
num_people = 4
tip_pct = 0.15

tip = bill_amount * tip_pct
total = bill_amount + tip
per_person = total / num_people

print(f"Bill: ₹{bill_amount} | Tip: {tip_pct:.0%} | People: {num_people}")
print(f"Tip amount: ₹{tip:.2f}")
print(f"Total: ₹{total:.2f}")
print(f"Per person: ₹{per_person:.2f}")
```

```python
# ANSWER 2: Grade Calculator
students = [("Rahul", 85), ("Priya", 92), ("Amit", 67), ("Neha", 45), ("Vikram", 78)]

for name, score in students:
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{name:8s} | Score: {score:3d} | Grade: {grade}")
```

```python
# ANSWER 3: Number Analyzer
daily_sales = [1200, 850, 2300, 0, 1750, 3200, 450, 0, 1100, 2800]

total = 0
non_zero_count = 0
zero_days = 0
max_sale = 0
min_sale = float('inf')
above_2000 = 0

for sale in daily_sales:
    total += sale
    if sale == 0:
        zero_days += 1
    else:
        non_zero_count += 1
        if sale > max_sale:
            max_sale = sale
        if sale < min_sale:
            min_sale = sale
    if sale > 2000:
        above_2000 += 1

avg = total / non_zero_count if non_zero_count > 0 else 0

print(f"Total sales: ₹{total:,}")
print(f"Average (excl. zero): ₹{avg:,.2f}")
print(f"Max sale: ₹{max_sale:,}")
print(f"Min sale (non-zero): ₹{min_sale:,}")
print(f"Zero-sale days: {zero_days}")
print(f"Days above ₹2000: {above_2000}")
```

```python
# ANSWER 5: Password Validator
passwords = ["Hello123", "short", "alllowercase1", "NOLOWER1", "NoDigits"]

for pwd in passwords:
    has_length = len(pwd) >= 8
    has_digit = False
    has_upper = False

    for char in pwd:
        if char.isdigit():
            has_digit = True
        if char.isupper():
            has_upper = True

    is_valid = has_length and has_digit and has_upper
    status = "✅ VALID" if is_valid else "❌ INVALID"

    reasons = []
    if not has_length:
        reasons.append("too short")
    if not has_digit:
        reasons.append("no digit")
    if not has_upper:
        reasons.append("no uppercase")

    detail = f" ({', '.join(reasons)})" if reasons else ""
    print(f"'{pwd}': {status}{detail}")
```
